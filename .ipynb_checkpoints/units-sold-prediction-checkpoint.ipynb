PREPROCESSING DATA

```python
# Print missing values
data.isnull().sum()
```

```python
# Print missing value from feature "Total Price"
data[data["Total Price"].isnull()]
```

```python
# Initiation mean feature "Total Price"
mean_total_price = data["Total Price"].mean()
# Print mean
mean_total_price
```

```python
# Add mean_total_price to missing value "Total Price"
data["Total Price"] = data["Total Price"].fillna(mean_total_price)
```

```python
# Print missing values feature "Total Pric"
data[data["Total Price"].isnull()]
```

```python
# Print index-136949 from feature "Total Price"
data[data["Total Price"].index == 136949]
```

```python
# Print dataset
data
```

```python
# Initiation feature
x = data[["Total Price", "Base Price"]] #predictors
y = data["Units Sold"] #prediction

# Split dataset
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42) #test_size = 0.2 artinya 20% dari dataset menjadi data test, random state agar hasil run konsisten
```

MODELING

```python
# Initiation model algorithm Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
# Process training data
model.fit(xtrain, ytrain)
```

EVALUATION

```python
# Try predict model with xtest
ypred = model.predict(xtest)
# Model evaluations
mse = mean_squared_error(ytest, ypred)  # MSE
rmse = root_mean_squared_error(ytest, ypred)  # RMSE

# Add matriks
mae = mae(ytest, ypred)
r2 = r2_score(ytest, ypred)

# Print Evaluation Results
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
```
